```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df
```

```python
x=df.drop(columns="price")
y=df["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train)
```

```python
y_pred=dtr.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

## Post Pruning 

```python
dtr=DecisionTreeRegressor(random_state=42)
```

```python
dtr.fit(x_train,y_train)
```

```python
y_pred=dtr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.figure(figsize=(16,8))
plot_tree(dtr,feature_names=x.columns,filled=True)
plt.show()
```

```python
path=dtr.cost_complexity_pruning_path(x_train,y_train)
```

```python
ccp_alphas=path.ccp_alphas
```

```python
ccp_alphas
```

```python
models=[]
for i in ccp_alphas:
    model=DecisionTreeRegressor(random_state=42,ccp_alpha=i)
    model.fit(x_train,y_train)
    models.append(model)
```

```python
models
```

```python
train_score=[]
test_score=[]
for model in models:
    y_train_pred=model.predict(x_train)
    y_test_pred=model.predict(x_test)
    train_score.append(r2_score(y_train,y_train_pred))
    test_score.append(r2_score(y_test,y_test_pred))
```

```python
train_score
```

```python
highest_index=np.argmax(test_score)
```

```python
max_score=np.max(test_score)
max_score
```

```python
best_alpha=ccp_alphas[highest_index]
best_alpha
```

```python
dtc=DecisionTreeRegressor(random_state=42,ccp_alpha=best_alpha)
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
```
